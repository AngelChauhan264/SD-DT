```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import make_regression
```

```python
 X,Y=make_regression(n_samples=10000,n_features=4,n_targets=1, noise=1000)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],Y)
```

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
from sklearn.tree import DecisionTreeRegressor
DTR=DecisionTreeRegressor()
```

```python
LR.fit(X,Y)
DTR.fit(X,Y)
```

```python
print(LR.score(X,Y))
print(DTR.score(X,Y))
```
